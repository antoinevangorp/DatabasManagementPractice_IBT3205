```python
import sqlite3

# Connect to a database (or create one if it doesn't exist)
conn = sqlite3.connect('your_database_name.db')
c = conn.cursor()

# Create tables
c.execute('''
    CREATE TABLE AircraftLocation (
        Aircrafts_id INTEGER NOT NULL,
        Airports_id INTEGER NOT NULL,
        PRIMARY KEY (Aircrafts_id, Airports_id),
        FOREIGN KEY (Aircrafts_id) REFERENCES Aircrafts (id),
        FOREIGN KEY (Airports_id) REFERENCES Airports (id)
    )
''')

c.execute('''
    CREATE TABLE Aircrafts (
        id INTEGER NOT NULL,
        type INTEGER NOT NULL,
        airline TEXT NOT NULL,
        capacity INTEGER NOT NULL,
        PRIMARY KEY (id)
    )
''')

c.execute('''
    CREATE TABLE Airports (
        id INTEGER NOT NULL,
        location TEXT NOT NULL,
        name TEXT NOT NULL,
        abbreviation TEXT NOT NULL,
        PRIMARY KEY (id)
    )
''')

c.execute('''
    CREATE TABLE Baggage (
        id INTEGER NOT NULL,
        weight INTEGER NOT NULL,
        status TEXT NOT NULL,
        Passengers_id INTEGER NOT NULL,
        Flights_id INTEGER NOT NULL,
        PRIMARY KEY (id),
        FOREIGN KEY (Passengers_id) REFERENCES Passengers (id),
        FOREIGN KEY (Flights_id) REFERENCES Flights (id)
    )
''')

c.execute('''
    CREATE TABLE Booking (
        id INTEGER NOT NULL,
        Flights_id INTEGER NOT NULL,
        Passengers_id INTEGER NOT NULL,
        seat TEXT NOT NULL,
        booking_date DATE NOT NULL,
        tiket_status TEXT NOT NULL,
        PRIMARY KEY (id),
        FOREIGN KEY (Flights_id) REFERENCES Flights (id),
        FOREIGN KEY (Passengers_id) REFERENCES Passengers (id)
    )
''')

c.execute('''
    CREATE TABLE CrewMembers (
        id INTEGER NOT NULL,
        FirstName TEXT NOT NULL,
        LastName TEXT NOT NULL,
        role TEXT NOT NULL,
        Aircrafts_id INTEGER NOT NULL,
        PRIMARY KEY (id),
        FOREIGN KEY (Aircrafts_id) REFERENCES Aircrafts (id)
    )
''')

c.execute('''
    CREATE TABLE Flights (
        id INTEGER NOT NULL,
        airline TEXT NOT NULL,
        departure TIMESTAMP NOT NULL,
        destination TIMESTAMP NOT NULL,
        Plains_id INTEGER NOT NULL,
        Airports_departure_id INTEGER NOT NULL,
        Airports_destination_id INTEGER NOT NULL,
        PRIMARY KEY (id),
        FOREIGN KEY (Plains_id) REFERENCES Aircrafts (id),
        FOREIGN KEY (Airports_departure_id) REFERENCES Airports (id),
        FOREIGN KEY (Airports_destination_id) REFERENCES Airports (id)
    )
''')

c.execute('''
    CREATE TABLE Passengers (
        id INTEGER NOT NULL,
        FirstName TEXT NOT NULL,
        Lastname TEXT NOT NULL,
        nationality TEXT NOT NULL,
        pasport TEXT,
        PRIMARY KEY (id)
    )
''')

# Create foreign key constraints
c.execute('''
    ALTER TABLE Booking ADD COLUMN Booking_Flights_id_fkey
    INTEGER REFERENCES Flights(id)
''')

c.execute('''
    ALTER TABLE Booking
    ADD COLUMN Booking_Passengers_id_fkey
    INTEGER REFERENCES Passengers(id)
''')

c.execute('''
    ALTER TABLE Flights
    ADD COLUMN Flights_Plains_id_fkey
    INTEGER REFERENCES Aircrafts(id)
''')

c.execute('''
    ALTER TABLE CrewMembers
    ADD COLUMN CrewMembers_Aircrafts_id_fkey
    INTEGER REFERENCES Aircrafts(id)
''')

c.execute('''
    ALTER TABLE Baggage
    ADD COLUMN Baggage_Passengers_id_fkey
    INTEGER REFERENCES Passengers(id)
''')

c.execute('''
    ALTER TABLE Baggage
    ADD COLUMN Baggage_Flights_id_fkey
    INTEGER REFERENCES Flights(id)
''')

c.execute('''
    ALTER TABLE Flights
    ADD COLUMN Flights_Airports_departure_id_fkey
    INTEGER REFERENCES Airports(id)
''')

c.execute('''
    ALTER TABLE Flights
    ADD COLUMN Flights_Airports_id_fkey1
    INTEGER REFERENCES Airports(id)
''')

c.execute('''
    ALTER TABLE AircraftLocation
    ADD COLUMN AircraftLocation_Aircrafts_id_fkey
    INTEGER REFERENCES Aircrafts(id)
''')

c.execute('''
    ALTER TABLE AircraftLocation
    ADD COLUMN AircraftLocation_Airports_id_fkey
    INTEGER REFERENCES Airports(id)
''')

# Commit the changes and close the connection
conn.commit()
conn.close()
```
